## The Euro Problem

*Information Theory, Inference, and Learning Algorithms* 에서 David MacKay 는 다음의 문제를 제안했다.

*The Guardian*  2002년 1월 4일 금요일에 다음과 같은 통계 기사가 게재되었다. :

> 벨기에 동전 1유로 동전을 축을 중심으로 250번 회전시켰는데 앞면이 140번, 뒷면이 110회가 나왔다 \ 런던 경제 학교 통계학 교수 배리 브라이트는 "매우 의심스럽다. 만약 동전이 한 쪽으로 기울어진 것이 아니라면, 결과가 이렇게 치우칠 확률은 7% 미만이다."

"하지만 이 데이터가 동전이 한 쪽으로 기울었다는 것의 증거가 될 수 있는가?"

이 질문에 대답하기 위해서는 두 단계를 거쳐야 한다.
1) 이항분포를 사용해 7%가 어디서 나왔는지를 찾아본다.
2) 베이즈 정리를 사용해 이 동전이 앞면이 나올 확률을 추정한다.

## The Binomial Distribution

만약 동전이 공정하다고 볼 수 있다면, 동전을 두 번 돌렸다면 `HH`, `HT`, `TH`, `TT` 4가지가 나올 수 있다. 각각의 결과는 25% 로 같다.
여기서 앞면이 나온 횟수를 더해보면 0, 1, or 2 이다.  0,2 는 25% 확률이고 1 is 50%이다.

앞면이 나올 확률을 $p$ 이고 이 동전을 $n$ 번 돌렸다고 가정해보자.  앞면이 총 $k$ 나올 확률은 이항분포를 따른다. [binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution):

$$\binom{n}{k} p^k (1-p)^{n-k}$$

$k$는 0 부터 $n$ 의 값을 포함한다.
$\binom{n}{k}$ 는 이항계수[binomial coefficient](https://en.wikipedia.org/wiki/Binomial_coefficient)라 부른다. (usually pronounced "n choose k").

이 분포는 직접 세서 구할 수 있지만, SciPy 함수 `binom.pmf` 를 사용해서 구할 수도 있다.
예를 들어 `n=2` 로 동전을 던졌을 때 앞면이 나올 확률이 `p=0.5` 라면 `k=1` 나올 확률은 아래와 같다:

```python
from scipy.stats import binom

n = 2
p = 0.5
k = 1

binom.pmf(k, n, p)
```

`binom.pmf` 함수에 단일값 `k` 대시 여러 값의 배열을 사용할 수도 있다.

```python
import numpy as np
ks = np.arange(n+1)
print(ks)
ps = binom.pmf(ks, n, p)
ps
```

결과로 0,1,2회의 앞면이 나올 확ㄹㄹ에 대한 numpy array 를 얻을 수 있다. 또한 `Pmf` 에 확률을 넣으면 `n` 과 `p` 에 대한 `k` 의 분포를 얻을 수 있다.

```python
from empiricaldist import Pmf

pmf_k = Pmf(ps, ks)
pmf_k
```

아래 함수는 주어진 `n` 과 `p` 에 대한 잉항분포를 구하여 이를 나타내는 `Pmf` 를 결괏값으로 반환하는 함수다.

```python
def make_binomial(n, p):
    """Make a binomial Pmf."""
    ks = np.arange(n+1)
    ps = binom.pmf(ks, n, p)
    return Pmf(ps, ks)
```

`n=250`, `p=0.5`인 이항분포는 아래와 같다.

```python
pmf_k = make_binomial(n=250, p=0.5)
```

```python
from utils import decorate

pmf_k.plot(label='n=250, p=0.5')

decorate(xlabel='Number of heads (k)',
         ylabel='PMF',
         title='Binomial distribution')
```

이 분포에서 가장 큰 확률을 가지는 값은 125다.

```python
pmf_k.max_prob()
```

하지만 확률이 가장 크다 하더라도 125회의 앞면이 나올 확률은 5% 정도밖에 안된다.

```python
pmf_k[125]
```

MacKay의 예시에서 140회의 앞면이 나올 확률은 더 작다.

```python
pmf_k[140]
```

동전이 한쪽으로 기울어진 것이 아니라면 결과가 치우쳐질 호가률이 7% 미만이라는 점을 아래의 함수로 확인이 가능하다.

```python
def prob_ge(pmf, threshold):
    """Probability of quantities greater than threshold."""
    ge = (pmf.qs >= threshold)
    total = pmf[ge].sum()
    return total
```

앞면이 140회 이상 나올 확률은 다음과 같다.

```python
prob_ge(pmf_k, 140)
```

`Pmf` 의 매서드를 활용해 동일한 결과를 얻을 수 있다.

```python
pmf_k.prob_ge(140)
```

결과는 3.3%로 언급했던 7%보다 더 낮은 확률이 도출된다. 통계학자는 140처럼 극단적인 모든 값을 언급했으므로, 110보다 같거나 작은 경우 역시 포함했기 때문에 이런 차이가 발생한다.
만약 기댓값인 125회보다 15회 많은 140 보다 큰 경우와 15회 적은 110회보다 적은 경우의 확률을 고려하면 7%정도가 도출된다.

```python
import matplotlib.pyplot as plt

def fill_below(pmf):
    qs = pmf.index
    ps = pmf.values
    plt.fill_between(qs, ps, 0, color='C5', alpha=0.4)

qs = pmf_k.index
fill_below(pmf_k[qs>=140])
fill_below(pmf_k[qs<=110])
pmf_k.plot(label='n=250, p=0.5')

decorate(xlabel='Number of heads (k)',
         ylabel='PMF',
         title='Binomial distribution')
```

Here's how we compute the total probability of the left tail.

```python
pmf_k.prob_le(110)
```

이 계산의 핵심은 코인이 공정하다고 한다면 극단적인 값은 나오지 않는다는 점이다.

## Bayesian Estimation
* 어떤 동전이든 세워서 돌리면 앞면을 보이며 넘어질 확률 `x` 를 가진다.
* 만약 동전의 무게가 완벽하게 균형을 이룬다면 `x`는 50%에 근접하리라 생각하겠지만 그렇지 않다면 무게에 따라 `x`은 달라진다.
* 이럴 때 베이즈 정리와 관측 데이터를 사용해서 `x`를 추정할 수 있다.

* 먼저 간단하게 `x`의 모든 값이 동일한 정도로 가능하다고 가정하는 균등분포를 사전분포를 두고 시작해보자. 균등분포는 numpy 로 정의한다.

```python
hypos = np.linspace(0, 1, 101)
prior = Pmf(1, hypos)
```

```python
prior
```

`hypos` 는 0~1사이의 값을 가지며 균등한 간격으로 구분된다. 이를 사용해 다음과 같이 가능도를 구한다.

```python
likelihood_heads = hypos
likelihood_tails = 1 - hypos
```

딕셔너리 구조로 앞면과 뒷면의 가능도를 표현한다.

```python
likelihood = {
    'H': likelihood_heads,
    'T': likelihood_tails
}
likelihood
```

데이터를 표현하기 위해 `H` 140회  `T` 110회 반복해 문자열을 만든다.

```python
dataset = 'H' * 140 + 'T' * 110
dataset
```

update 를 위해 아래의 함수를 활용한다.

```python
def update_euro(pmf, dataset):
    """Update pmf with a given sequence of H and T."""
    for data in dataset:
        pmf *= likelihood[data]

    pmf.normalize()
```

* 함수 내 반복문이 반복되면서 pmf 에 문자열의 결과값에 가능도를 곱해 나간다.
* normalize 가 반복문 밖에 있는데, 사후확률분포는 끝에 한번만 표준화해준다. 동전을 돌리고 표준화하는 것보다 훨씬 효율적이다.

```python
posterior = prior.copy()
update_euro(posterior, dataset)
```

```python
posterior
```

사후확률분포는 아래와 같다.

```python
def decorate_euro(title):
    decorate(xlabel='Proportion of heads (x)',
             ylabel='Probability',
             title=title)
```

```python
posterior.plot(label='140 heads out of 250', color='C4')
decorate_euro(title='Posterior distribution of x')
```

* 사후확률분포의 `x` 에 해당하는 부분은 앞면의 비율이다.
* 사후 확률분포는 데이터를 본 이후의 `x` 에 대한 믿음을 나타낸다.
* 이 그림을 보면 0.4보다 작거나 0.7보다 큰 경우는 거의 없고 0.5~0.6 사이가 가장 가능성이 있음을 말해준다.
* 가장 가능성 있는 `x` 는 0.56 이고 이는 횟수의 비율로 나타내면  `140/250` 이다.

```python
posterior.max_prob()
```

## Triangle Prior

So far we've been using a uniform prior:

```python
uniform = Pmf(1, hypos, name='uniform')
uniform.normalize()
uniform
```

* 앞서 균등분포를 사전확률분포로 활용했는데 이는 우리가 동전에 대해 알고 있는 지식을 생각해보면 그다지 합리적이지 않다.
* 동전의 무게가 한쪽으로 쏠려있다면 `x` 는 0.5에서 살짝 벗어나있을수 있고 0.1이나 0.9와 같이 극단적인 값은 거의 나오지 않을것 이다.

* 따라서 `x` 값이 0.5에 가까울수록 높고, 극단값에서는 낮은 확률을 가지는 사전분포를 선택하는게 합리적이다.
* 이 예시로 삼각형 형태의 사전확률을 고려해볼 수 있다.

```python
ramp_up = np.arange(50)
ramp_down = np.arange(50, -1, -1)

a = np.append(ramp_up, ramp_down)

triangle = Pmf(a, hypos, name='triangle')
triangle.normalize()
```

`arange` 는 numpy array를 반환하고 `np.append` 를 이용해 `ramp_down` 와 `ramp_up`를 이어붙인다.
아래 plot 은 이 사전분포의 형태를 보여준다.

```python
uniform.plot()
triangle.plot()
decorate_euro(title='Uniform and triangle prior distributions')
```

Now we can update both priors with the same data:


```python
update_euro(uniform, dataset)
update_euro(triangle, dataset)
```

Here are the posteriors.

```python
uniform.plot()
triangle.plot()
decorate_euro(title='Posterior distributions')
```

* 두 사후확률분포는 눈으로 확인했을 때 큰 차이가 없다. 즉, 균등분포와 삼각분포 중 어느 사전확률분포를 쓰더라도 데이터로 믿음을 udpate 하게 되면 유사한 사후확률분포를 얻을 수 있다는 결론에 이른다.
* 데이터가 충분하다면 서로 다른 사전확률로 시작한다고 해도 동일한 사후확률로 수렴하는 경향이 있다.

## The Binomial Likelihood Function

* 위에서 동전을 한번 돌릴때 마다 확률을 갱신했으므로 유로 동전 문제에서는 갱신을 250하게 된다.
* 좀 더 효율적으로 데이터셋에 대한 가능도를 계산하기 위해서는 이 사례에서는 이항분포를 활용하면 된다.
앞면이 나올 확률이 $p$ 일때, $n$ 의 총 시행중 앞면이 $k$ 번 나올 확률은 아래와 같다.

$$\binom{n}{k} p^k (1-p)^{n-k}$$

* Scipy 를 활용해 이항분포를 활용할 수 있다. 구해진 가능도에 사전확률을 곱해 update 하는 과정을 거지면 된다.

```python
from scipy.stats import binom

def update_binomial(pmf, data):
    """Update pmf using the binomial distribution."""
    k, n = data
    xs = pmf.qs
    likelihood = binom.pmf(k, n, xs)
    pmf *= likelihood
    pmf.normalize()
```

데이터는 문자열 대신 `k` 와 `n`을 선언해준다.

```python
uniform2 = Pmf(1, hypos, name='uniform2')
data = 140, 250
update_binomial(uniform2, data)
```

And here's what the posterior looks like.

```python
uniform.plot()
uniform2.plot()
decorate_euro(title='Posterior distributions computed two ways')
```

`allclose` 를 활용해 매우 작은 소숫점 이하 부분은 반올림한 후 앞에서 구한 결과와 같은지를 확인할 수 있다.

```python
np.allclose(uniform, uniform2)
```

## Bayesian Statistics

* 이전 장에서 본 101개 쿠키 그릇 문제와 이 유로 동전문제는 사전분포도 동일하고 가능도도 동일하고 데이터도 같아 결과 역시 같다. 하지만 두 가지 차이점이 있다.
* 먼저 사전분포의 선택이 다른데, 쿠키 그릇 문제에서는 그릇 하나를 동일한 확률로 고른다는 조건으로 균등분포가 문제에서 주어졋으나, 유로 동전 문제는 사전분포를 주관적으로 선택할 수 있다.
* 사전분포가 주관적이니 사후분포 역시 주관적이라 이 방법에 문제를 제기할 수 있다.
* 두 번째로 추정방식에 차이가 있다. 쿠키 그릇 문제에서는 그릇을 임의 로 선택하므로 각 그릇을 선택하는 확률을 구하는 것은 모순적이지 않다. 반면 앞면의 비율은 동전의 물리적 속성인데 특정 확률 해석의 경우 물리적 속성은 임의로 고려하는 것이 문제가 되기도 한다.
    * 베이지안 방식의 확률 해석에 있어, 물리량은 임의 상태로 두고 이에 대한 확률을 계산하는 것은 의미 있고 유용하다 할 수 있다.
* 베이즈의 정리는 확률에 대한 수학 공식이므로 이떤 이견도 없는 공식이나, 베이지안 통계는 논란이 되는 부분도 있다.

## Summary
* 유로 동전 문제를 소개하고 이를 데이터로 표현해 앞면이 나올 확률 $x$ 의 사후확률을 구해보았다.
* 두 가지 다른 사전확률을 만들고 동일한 데이터로 사후확률분포를 도출해냈다. 그 결과로 거의 동일한 두 분포를 확인할 수 있었다.
    * 서로 다른 belief 가 같은 데이터로 시작하면 belief는 수렴할 수 있다.
* 사후확률분포를 보다 효율적으로 구하기 위해 이항분포를 활용해보았다.
* 하지만 데이터가 동전이 한쪽으로 기울었다는 증거가 되었는지에 대해 답하진 않았다.

Knowing this, people might be more willing to answer honestly.

Suppose you survey 100 people this way and get 80 YESes and 20 NOs.  Based on this data, what is the posterior distribution for the fraction of people who cheat on their taxes?  What is the most likely quantity in the posterior distribution?

```python
# Solution goes here
# 탈세자의 비율 x 가 균등분포를 따른다고 가정
hypos = np.linspace(0, 1, 101)
prior = Pmf(1, hypos)
```

```python
# 가능도 계산
# 먼저 No 라고 대답한 비율은 탈세자가 아닌 사람들만 대답할 수 있고 뒷면만 나오는 경우에만 대답이 가능하기 때문에, 1-탈세자의 비율 / 동전이 뒷면 나오는 경우 (1/2) 를 곱한값이 된다.
# 'Yes' 라고 대답하는 경우는 동전이 앞면이 나오는 경우 무조건 대답하기 때문에 0.5 를 더하고, 탈세자 중 절반의 비율로 뒷면이 나오고 예라고 답할 가능성이라 한다.
likelihood = {
    'Y' : 0.5 + hypos/2,
    'N' : (1-hypos)/2
}
```

```python
dataset = 'Y' * 80 + 'N' * 20
for data in dataset:
    prior *= likelihood[data]

prior.normalize()
posterior = prior.copy()
```

```python
# distribution plot
posterior.plot(label='80 YES, 20 NO')
decorate(xlabel='ratio of cheaters',
         ylabel='PMF')
```

```python
# the most likely quantity in the posterior distribution
posterior.max_prob()
```

**Exercise:** Suppose you want to test whether a coin is fair, but you don't want to spin it hundreds of times.
So you make a machine that spins the coin automatically and uses computer vision to determine the outcome.

However, you discover that the machine is not always accurate.  Specifically, suppose the probability is `y=0.2` that an actual heads is reported as tails, or actual tails reported as heads.

If we spin a coin 250 times and the machine reports 140 heads, what is the posterior distribution of `x`?
What happens as you vary the value of `y`?

```python
# Solution goes here
# 역시 가능도를 계산하는 부분을 잘 반영해야한다.
# 실제 앞면을 뒷면으로 기록할 확률 or 실제 뒷면을 앞면이라고 기록할 확률 y
# 1-y : 실제 앞면을 앞면이라고 기록할 확률 or 실제 뒷면을 뒷면이라고 기록할 확률
# 전체 가능도 
# 앞면 : (1-y)*실제 앞면의 비율 + y*실제 뒷면의 비율
# 뒷면 : (1-y)*실제 뒷면의 비율 + 


def update_unreliable(pmf, dataset, y):
    likelihood = {
        'H': (1-y) * hypos + y * (1-hypos),
        'T': y * hypos + (1-y) * (1-hypos)
    }
    for data in dataset:
        pmf *= likelihood[data]
    pmf.normalize()
```

```python
hypos = np.linspace(0, 1, 101)
prior = Pmf(1, hypos)
dataset = "H"*140 + "T"*110
update_unreliable(prior, dataset, y=0.2)
```

```python
# Solution goes here
posterior = prior.copy()
posterior.plot()
decorate(xlabel='ratio of head',
         ylabel='PMF')
```

```python
def test_y_posterior(y_val):
    hypos = np.linspace(0, 1, 101)
    prior = Pmf(1, hypos)
    dataset = "H"*140 + "T"*110
    update_unreliable(prior, dataset, y=y_val)
    return prior
```

```python
# Solution goes here
y_list = [0.01, 0.1, 0.2, 0.3, 0.5]
posterior_list = [test_y_posterior(y) for y in y_list]
for i,y in enumerate(y_list):
    posterior_list[i].plot(label = "y = {}".format(y))
decorate(xlabel='ratio of head',ylabel='PMF')
```

**Exercise:** In preparation for an alien invasion, the Earth Defense League (EDL) has been working on new missiles to shoot down space invaders.  Of course, some missile designs are better than others; let's assume that each design has some probability of hitting an alien ship, `x`.

Based on previous tests, the distribution of `x` in the population of designs is approximately uniform between 0.1 and 0.4.

Now suppose the new ultra-secret Alien Blaster 9000 is being tested.  In a press conference, an EDL general reports that the new design has been tested twice, taking two shots during each test.  The results of the test are confidential, so the general won't say how many targets were hit, but they report: "The same number of targets were hit in the two tests, so we have reason to think this new design is consistent."

Is this data good or bad? 
That is, does it increase or decrease your estimate of `x` for the Alien Blaster 9000?

Hint: If the probability of hitting each target is $x$, the probability of hitting one target in both tests
is $\left[2x(1-x)\right]^2$.

```python
# Solution goes here
hypos = np.linspace(0.1, 0.4, 101)
prior = Pmf(1, hypos)
prior
```

```python
hypos
```

```python
# 동일한 수의 목표물이 격추되었다-> [[0,0] ,[1,1] ,[2,2]]
# 위의 경우의 수에 대한 확률을 모두 더해야 likelihood 가 도출됨
x = hypos
hits = [(1-x)**4, (2*x*(1-x))**2, x**4]
likelihood = np.sum(hits, axis= 0) 
likelihood
```

```python
n = 2
likes2 = [binom.pmf(k, n, x)**2 for k in range(n+1)]
likelihood2 = np.sum(likes2, axis=0)
```

```python
plt.plot(x, likelihood, label='special case')
plt.plot(x, likelihood2, label='general formula')
decorate(xlabel='Probability of hitting the target',
         ylabel='Likelihood',
         title='Likelihood of getting the same result')
```

```python
posterior = prior * likelihood
posterior.normalize()
```

```python
posterior.plot(label='Two tests, two shots, same outcome',
               color='C4')
decorate(xlabel='Probability of hitting the target',
         ylabel='PMF',
         title='Posterior distribution',
         ylim=[0, 0.015])
```

```python
# Getting the same result in both tests is more likely for extreme values of `x` and least likely when `x=0.5`.

# In this example, the prior indicates that `x` is less than 0.5,and the update gives more weight to extreme values.

# So the dataset makes lower values of `x` more likely.
```
